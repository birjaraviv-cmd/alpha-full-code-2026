```python
import numpy as np
import matplotlib.pyplot as plt
from typing import List
from dataclasses import dataclass
from abc import ABC, abstractmethod
import time
from datetime import timedelta


plt.rcParams['font.family'] = 'DejaVu Sans'

# ╔══════════════════════════════════════════════════════════════════════════╗
# ║ 🎯 הגדרות - שנה רק כאן! 🎯 ║
# ╚══════════════════════════════════════════════════════════════════════════╝

NUM_VOTERS = 100
NUM_ROUNDS = 50
NUM_CANDIDATES = 300
PROJECT_COUNTS = [2, 3, 4, 5]
REPEATS_PER_DISTRIBUTION = 10
BASE_SEED = 42


@dataclass
class Voter:
    id: int
    error_rate: float
    weight: float = 1.0
    previous_vote: np.ndarray = None

class WeightUpdateStrategy(ABC):
    @abstractmethod
    def update_weights_batch(self, voters: List[Voter], votes: np.ndarray, current_result: np.ndarray, round_num: int) -> None:
        pass
    @abstractmethod
    def get_name(self) -> str:
        pass

# ════════════════════════════════════════════════════════════════════════════
# 10 Linear Ranking Strategies
# ════════════════════════════════════════════════════════════════════════════

class LinearRankingRewardStrategy(WeightUpdateStrategy):
    def __init__(self, max_bonus_percent: float = 10.0):
        self.max_bonus_percent = max_bonus_percent
    def update_weights_batch(self, voters, votes, current_result, round_num):
        distances = [(np.linalg.norm(votes[i] - current_result), i, voters[i]) for i in range(len(voters))]
        distances.sort(key=lambda x: x[0])
        num_voters = len(voters)
        for rank, (dist, idx, voter) in enumerate(distances):
            if num_voters > 1:
                bonus_percent = self.max_bonus_percent * (num_voters - rank - 1) / (num_voters - 1)
            else:
                bonus_percent = self.max_bonus_percent
            multiplier = 1.0 + (bonus_percent / 100.0)
            voter.weight = voter.weight * multiplier
            voter.weight = max(0.01, min(100.0, voter.weight))
    def get_name(self):
        return f"Linear Ranking {self.max_bonus_percent:.0f}%"

class LinearRanking10Strategy(LinearRankingRewardStrategy):
    def __init__(self): super().__init__(10.0)
class LinearRanking20Strategy(LinearRankingRewardStrategy):
    def __init__(self): super().__init__(20.0)
class LinearRanking30Strategy(LinearRankingRewardStrategy):
    def __init__(self): super().__init__(30.0)
class LinearRanking40Strategy(LinearRankingRewardStrategy):
    def __init__(self): super().__init__(40.0)
class LinearRanking50Strategy(LinearRankingRewardStrategy):
    def __init__(self): super().__init__(50.0)
class LinearRanking60Strategy(LinearRankingRewardStrategy):
    def __init__(self): super().__init__(60.0)
class LinearRanking70Strategy(LinearRankingRewardStrategy):
    def __init__(self): super().__init__(70.0)
class LinearRanking80Strategy(LinearRankingRewardStrategy):
    def __init__(self): super().__init__(80.0)
class LinearRanking90Strategy(LinearRankingRewardStrategy):
    def __init__(self): super().__init__(90.0)
class LinearRanking100Strategy(LinearRankingRewardStrategy):
    def __init__(self): super().__init__(100.0)

# ════════════════════════════════════════════════════════════════════════════
# Voting System
# ════════════════════════════════════════════════════════════════════════════

class VotingSystem:
    def __init__(self, num_voters: int, weight_strategy: WeightUpdateStrategy, num_projects: int=3,
                 num_candidates: int=1000, random_seed: int=None, error_distribution: str='uniform'):
        if random_seed is not None:
            np.random.seed(random_seed)
        self.num_voters = num_voters
        self.num_projects = num_projects
        self.num_candidates = num_candidates
        self.true_allocation = self._generate_random_truth()
        self.weight_strategy = weight_strategy
        self.error_distribution = error_distribution
        self.shared_bias_direction = None
        if error_distribution == 'shared_bias':
            self.shared_bias_direction = self._generate_shared_bias()
        self.voters = self._initialize_voters()
        self.error_history = []

    def _generate_random_truth(self):
        vals = np.random.random(self.num_projects)
        return vals/np.sum(vals)

    def _generate_shared_bias(self):
        bias = np.random.normal(0, 1, self.num_projects)
        return bias / np.linalg.norm(bias)

    def _initialize_voters(self):
        voters = []
        if self.error_distribution == 'uniform':
            for i in range(self.num_voters):
                voters.append(Voter(i, np.random.uniform(0, 2)))
        elif self.error_distribution == 'good_majority':
            num_good = int(0.7 * self.num_voters)
            for i in range(num_good):
                voters.append(Voter(i, np.random.uniform(0, 0.5)))
            for i in range(num_good, self.num_voters):
                voters.append(Voter(i, np.random.uniform(0.5, 2.0)))
        elif self.error_distribution == 'bad_majority':
            num_bad = int(0.7 * self.num_voters)
            for i in range(num_bad):
                voters.append(Voter(i, np.random.uniform(1.5, 2.0)))
            for i in range(num_bad, self.num_voters):
                voters.append(Voter(i, np.random.uniform(0, 1.5)))
        elif self.error_distribution == 'shared_bias':
            for i in range(self.num_voters):
                voters.append(Voter(i, np.random.uniform(0, 2)))
        return voters

    def generate_vote_from_truth(self, voter: Voter, round_num: int):
        np.random.seed(voter.id * 12345 + round_num)
        if self.error_distribution == 'shared_bias':
            error_vector = self.shared_bias_direction * voter.error_rate
            vote = self.true_allocation + error_vector
            vote = np.abs(vote)
            return vote / np.sum(vote)
        if voter.previous_vote is None:
            rand = np.random.random(self.num_projects)
            rand = rand / np.sum(rand)
            mix_factor = voter.error_rate / 2.0
            vote = (1 - mix_factor) * self.true_allocation + mix_factor * rand
            vote = vote / np.sum(vote)
            voter.previous_vote = vote
            return vote
        candidates = []
        for i in range(self.num_candidates):
            rand = np.random.random(self.num_projects)
            rand = rand / np.sum(rand)
            mix_factor = voter.error_rate / 2.0
            vote_candidate = (1 - mix_factor) * self.true_allocation + mix_factor * rand
            candidates.append(vote_candidate / np.sum(vote_candidate))
        candidates = np.array(candidates)
        distances = np.array([np.sum(np.abs(candidate - voter.previous_vote)) for candidate in candidates])
        weights = 1.0 / (distances + 1e-10)
        probabilities = weights / np.sum(weights)
        idx = np.random.choice(len(candidates), p=probabilities)
        vote = candidates[idx].copy()
        voter.previous_vote = vote
        return vote

    def aggregate_votes(self, votes: np.ndarray, weights: np.ndarray):
        return np.sum(votes*weights[:,None], axis=0)/np.sum(weights)

    def run_voting_round(self, round_num: int):
        votes = np.array([self.generate_vote_from_truth(voter, round_num) for voter in self.voters])
        weights = np.array([voter.weight for voter in self.voters])
        result = self.aggregate_votes(votes, weights)
        error = np.sum(np.abs(result - self.true_allocation))
        self.weight_strategy.update_weights_batch(self.voters, votes, result, round_num)
        self.error_history.append(error)
        return error

    def run_multiple_rounds(self, num_rounds: int, verbose: bool = True):
        for round_num in range(num_rounds):
            self.run_voting_round(round_num)
        return self.error_history

# ════════════════════════════════════════════════════════════════════════════
# Run Single Algorithm
# ════════════════════════════════════════════════════════════════════════════

def run_single_algorithm_with_repeats(strategy_class, num_voters, num_rounds, num_projects, num_candidates, base_seed, repeats_per_dist):
    distributions = ['uniform', 'good_majority', 'bad_majority', 'shared_bias']
    results_by_distribution = {}
    raw_errors_by_distribution = {}
    for dist_idx, dist_name in enumerate(distributions):
        dist_errors = []
        for run in range(repeats_per_dist):
            strategy = strategy_class()
            seed = base_seed + dist_idx * 10000 + run * 1000
            system = VotingSystem(num_voters, strategy, num_projects, num_candidates, seed, error_distribution=dist_name)
            errors = system.run_multiple_rounds(num_rounds, verbose=False)
            dist_errors.append(errors)
        dist_errors = np.array(dist_errors)
        raw_errors_by_distribution[dist_name] = dist_errors
        avg_dist_errors = np.mean(dist_errors, axis=0)
        initial = avg_dist_errors[0]
        final = avg_dist_errors[-1]
        minimum = np.min(avg_dist_errors)
        improvement = initial - final
        improvement_percent = (improvement / initial * 100) if initial > 0 else 0
        results_by_distribution[dist_name] = {'initial': initial, 'final': final, 'min': minimum, 'improvement': improvement, 'improvement_percent': improvement_percent}
    avg_final = np.mean([results_by_distribution[dist]['final'] for dist in distributions])
    avg_initial = np.mean([results_by_distribution[dist]['initial'] for dist in distributions])
    avg_improvement_percent = np.mean([results_by_distribution[dist]['improvement_percent'] for dist in distributions])
    avg_min = np.mean([results_by_distribution[dist]['min'] for dist in distributions])
    return {'name': strategy_class().get_name(), 'final': avg_final, 'initial': avg_initial, 'min': avg_min, 'improvement': avg_initial - avg_final, 'improvement_percent': avg_improvement_percent, 'by_distribution': results_by_distribution,'raw_errors': raw_errors_by_distribution
}

# ════════════════════════════════════════════════════════════════════════════
# Plot Function - One plot per algorithm (averaged across all project counts)
# ════════════════════════════════════════════════════════════════════════════

def plot_error_over_rounds_aggregated(strategy_name, all_raw_errors_by_project):
    """
    all_raw_errors_by_project:
    dict -> num_projects : dict -> distribution name : np.array shape (repeats, num_rounds)
    """

    distributions = ['uniform', 'good_majority', 'bad_majority', 'shared_bias']
    num_rounds = all_raw_errors_by_project[PROJECT_COUNTS[0]][distributions[0]].shape[1]
    rounds = np.arange(num_rounds)

    plt.figure(figsize=(10, 6))

    # Collect all errors across all project counts for each distribution
    all_means = []
    for dist in distributions:
        # Average across all project counts for this distribution
        dist_means_across_projects = []
        for num_projects in PROJECT_COUNTS:
            mean_error = np.mean(all_raw_errors_by_project[num_projects][dist], axis=0)
            dist_means_across_projects.append(mean_error)

        # Average across all project counts
        avg_dist_mean = np.mean(np.array(dist_means_across_projects), axis=0)
        all_means.append(avg_dist_mean)
        plt.plot(rounds, avg_dist_mean, label=dist)

    # Global average across all distributions and project counts
    global_mean = np.mean(np.array(all_means), axis=0)
    plt.plot(rounds, global_mean, linewidth=3, linestyle='--', label='Global Average', color='black')

    plt.title(f'Error vs Voting Round – {strategy_name}\n(Averaged across {len(PROJECT_COUNTS)} project counts: {PROJECT_COUNTS})')
    plt.xlabel('Voting Round')
    plt.ylabel('Error')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

# ════════════════════════════════════════════════════════════════════════════
# Run Linear Ranking Algorithms - MAIN EXECUTION
# ════════════════════════════════════════════════════════════════════════════

all_strategies = [
    LinearRanking10Strategy, LinearRanking20Strategy, LinearRanking30Strategy,
    LinearRanking40Strategy, LinearRanking50Strategy, LinearRanking60Strategy,
    LinearRanking70Strategy, LinearRanking80Strategy, LinearRanking90Strategy,
    LinearRanking100Strategy
]

total_algorithms = len(all_strategies)
total_runs = total_algorithms * len(PROJECT_COUNTS)
total_simulations = total_runs * REPEATS_PER_DISTRIBUTION * 4

print(f"\n{'='*130}")
print(f"🚀 LINEAR RANKING ALGORITHMS")
print(f"🚀 Running {total_algorithms} algorithms × {len(PROJECT_COUNTS)} project counts × 4 distributions × {REPEATS_PER_DISTRIBUTION} repeats")
print(f"   📊 Total: {total_runs} main runs ({total_simulations} simulations!)")
print(f"   📊 Each simulation: {NUM_VOTERS} voters, {NUM_ROUNDS} rounds, {NUM_CANDIDATES} candidates")
print(f"   📈 Will create {total_algorithms} plots (one per algorithm, averaged across all project counts)")
print(f"{'='*130}\n")

start_total = time.time()
results_by_algorithm = {}
completed = 0

for algo_idx, strategy_class in enumerate(all_strategies, 1):
    strategy_name = strategy_class().get_name()
    results_by_algorithm[strategy_name] = {}

    # Store raw errors for all project counts
    all_raw_errors_by_project = {}

    print(f"\n┌{'─'*126}┐")
    print(f"│ [{algo_idx}/{total_algorithms}] 🔄 {strategy_name:<116} │")
    print(f"└{'─'*126}┘")

    for proj_idx, num_projects in enumerate(PROJECT_COUNTS, 1):
        start = time.time()
        print(f"   └─ [{proj_idx}/{len(PROJECT_COUNTS)}] {num_projects} Projects... ", end='', flush=True)

        results = run_single_algorithm_with_repeats(strategy_class, NUM_VOTERS, NUM_ROUNDS, num_projects, NUM_CANDIDATES, BASE_SEED, REPEATS_PER_DISTRIBUTION)
        results_by_algorithm[strategy_name][num_projects] = results

        # Store raw errors for later aggregation
        all_raw_errors_by_project[num_projects] = results['raw_errors']

        elapsed = time.time() - start
        completed += 1
        avg_time_per_run = (time.time() - start_total) / completed
        remaining_runs = total_runs - completed
        eta_seconds = avg_time_per_run * remaining_runs
        eta = str(timedelta(seconds=int(eta_seconds)))
        progress_pct = (completed / total_runs) * 100

        dist_results = [f"{dist[:3].upper()}:{results['by_distribution'][dist]['improvement_percent']:>5.1f}%" for dist in ['uniform', 'good_majority', 'bad_majority', 'shared_bias']]
        dist_str = " | ".join(dist_results)

        print(f"✅ {elapsed:>5.1f}s")
        print(f"      📊 Avg Error: {results['final']:.4f} | Avg Improvement: {results['improvement_percent']:>5.1f}%")
        print(f"      📈 By Distribution: {dist_str}")
        print(f"      ⏱️  Progress: {progress_pct:>5.1f}% | ETA: {eta}")

    # After all project counts are done, create ONE aggregated plot for this algorithm
    print(f"   📊 Creating aggregated plot for {strategy_name}...")
    plot_error_over_rounds_aggregated(strategy_name, all_raw_errors_by_project)


total_time = time.time() - start_total
print(f"\n{'='*130}")
print(f"✅ LINEAR RANKING ALGORITHMS COMPLETED! Total Time: {str(timedelta(seconds=int(total_time)))}")
print(f"✅ Created {total_algorithms} plots (one per algorithm)")
print(f"{'='*130}\n")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from typing import List
from dataclasses import dataclass
from abc import ABC, abstractmethod
import time
from datetime import timedelta


plt.rcParams['font.family'] = 'DejaVu Sans'

# ╔══════════════════════════════════════════════════════════════════════════╗
# ║ 🎯 הגדרות - שנה רק כאן! 🎯 ║
# ╚══════════════════════════════════════════════════════════════════════════╝

NUM_VOTERS = 100
NUM_ROUNDS = 50
NUM_CANDIDATES = 300
PROJECT_COUNTS = [2, 3, 4, 5]
REPEATS_PER_DISTRIBUTION = 10
BASE_SEED = 42


@dataclass
class Voter:
    id: int
    error_rate: float
    weight: float = 1.0
    previous_vote: np.ndarray = None

class WeightUpdateStrategy(ABC):
    @abstractmethod
    def update_weights_batch(self, voters: List[Voter], votes: np.ndarray, current_result: np.ndarray, round_num: int) -> None:
        pass
    @abstractmethod
    def get_name(self) -> str:
        pass

# ════════════════════════════════════════════════════════════════════════════
# 9 Half Split Strategies
# ════════════════════════════════════════════════════════════════════════════

class HalfSplitDynamicStrategy(WeightUpdateStrategy):
    def __init__(self, mode: int):
        self.mode = mode
        self.top_multiplier = float(mode + 1)
        self.bottom_multiplier = 1.0 / float(mode + 1)
    def update_weights_batch(self, voters, votes, current_result, round_num):
        distances = [(np.linalg.norm(votes[i] - current_result), i) for i in range(len(voters))]
        distances.sort(key=lambda x: x[0])
        n = len(voters)
        half = n // 2
        for rank, (_, idx) in enumerate(distances):
            voter = voters[idx]
            if rank < half:
                mult = self.top_multiplier
            else:
                mult = self.bottom_multiplier
            voter.weight *= mult
            voter.weight = float(np.clip(voter.weight, 0.01, 100.0))
    def get_name(self):
        return f"Half Split Mode {self.mode}"

class HalfSplitMode1Strategy(HalfSplitDynamicStrategy):
    def __init__(self): super().__init__(1)
class HalfSplitMode2Strategy(HalfSplitDynamicStrategy):
    def __init__(self): super().__init__(2)
class HalfSplitMode3Strategy(HalfSplitDynamicStrategy):
    def __init__(self): super().__init__(3)
class HalfSplitMode4Strategy(HalfSplitDynamicStrategy):
    def __init__(self): super().__init__(4)
class HalfSplitMode5Strategy(HalfSplitDynamicStrategy):
    def __init__(self): super().__init__(5)
class HalfSplitMode6Strategy(HalfSplitDynamicStrategy):
    def __init__(self): super().__init__(6)
class HalfSplitMode7Strategy(HalfSplitDynamicStrategy):
    def __init__(self): super().__init__(7)
class HalfSplitMode8Strategy(HalfSplitDynamicStrategy):
    def __init__(self): super().__init__(8)
class HalfSplitMode9Strategy(HalfSplitDynamicStrategy):
    def __init__(self): super().__init__(9)

# ════════════════════════════════════════════════════════════════════════════
# Voting System
# ════════════════════════════════════════════════════════════════════════════

class VotingSystem:
    def __init__(self, num_voters: int, weight_strategy: WeightUpdateStrategy, num_projects: int=3,
                 num_candidates: int=1000, random_seed: int=None, error_distribution: str='uniform'):
        if random_seed is not None:
            np.random.seed(random_seed)
        self.num_voters = num_voters
        self.num_projects = num_projects
        self.num_candidates = num_candidates
        self.true_allocation = self._generate_random_truth()
        self.weight_strategy = weight_strategy
        self.error_distribution = error_distribution
        self.shared_bias_direction = None
        if error_distribution == 'shared_bias':
            self.shared_bias_direction = self._generate_shared_bias()
        self.voters = self._initialize_voters()
        self.error_history = []

    def _generate_random_truth(self):
        vals = np.random.random(self.num_projects)
        return vals/np.sum(vals)

    def _generate_shared_bias(self):
        bias = np.random.normal(0, 1, self.num_projects)
        return bias / np.linalg.norm(bias)

    def _initialize_voters(self):
        voters = []
        if self.error_distribution == 'uniform':
            for i in range(self.num_voters):
                voters.append(Voter(i, np.random.uniform(0, 2)))
        elif self.error_distribution == 'good_majority':
            num_good = int(0.7 * self.num_voters)
            for i in range(num_good):
                voters.append(Voter(i, np.random.uniform(0, 0.5)))
            for i in range(num_good, self.num_voters):
                voters.append(Voter(i, np.random.uniform(0.5, 2.0)))
        elif self.error_distribution == 'bad_majority':
            num_bad = int(0.7 * self.num_voters)
            for i in range(num_bad):
                voters.append(Voter(i, np.random.uniform(1.5, 2.0)))
            for i in range(num_bad, self.num_voters):
                voters.append(Voter(i, np.random.uniform(0, 1.5)))
        elif self.error_distribution == 'shared_bias':
            for i in range(self.num_voters):
                voters.append(Voter(i, np.random.uniform(0, 2)))
        return voters

    def generate_vote_from_truth(self, voter: Voter, round_num: int):
        np.random.seed(voter.id * 12345 + round_num)
        if self.error_distribution == 'shared_bias':
            error_vector = self.shared_bias_direction * voter.error_rate
            vote = self.true_allocation + error_vector
            vote = np.abs(vote)
            return vote / np.sum(vote)
        if voter.previous_vote is None:
            rand = np.random.random(self.num_projects)
            rand = rand / np.sum(rand)
            mix_factor = voter.error_rate / 2.0
            vote = (1 - mix_factor) * self.true_allocation + mix_factor * rand
            vote = vote / np.sum(vote)
            voter.previous_vote = vote
            return vote
        candidates = []
        for i in range(self.num_candidates):
            rand = np.random.random(self.num_projects)
            rand = rand / np.sum(rand)
            mix_factor = voter.error_rate / 2.0
            vote_candidate = (1 - mix_factor) * self.true_allocation + mix_factor * rand
            candidates.append(vote_candidate / np.sum(vote_candidate))
        candidates = np.array(candidates)
        distances = np.array([np.sum(np.abs(candidate - voter.previous_vote)) for candidate in candidates])
        weights = 1.0 / (distances + 1e-10)
        probabilities = weights / np.sum(weights)
        idx = np.random.choice(len(candidates), p=probabilities)
        vote = candidates[idx].copy()
        voter.previous_vote = vote
        return vote

    def aggregate_votes(self, votes: np.ndarray, weights: np.ndarray):
        return np.sum(votes*weights[:,None], axis=0)/np.sum(weights)

    def run_voting_round(self, round_num: int):
        votes = np.array([self.generate_vote_from_truth(voter, round_num) for voter in self.voters])
        weights = np.array([voter.weight for voter in self.voters])
        result = self.aggregate_votes(votes, weights)
        error = np.sum(np.abs(result - self.true_allocation))
        self.weight_strategy.update_weights_batch(self.voters, votes, result, round_num)
        self.error_history.append(error)
        return error

    def run_multiple_rounds(self, num_rounds: int, verbose: bool = True):
        for round_num in range(num_rounds):
            self.run_voting_round(round_num)
        return self.error_history

# ════════════════════════════════════════════════════════════════════════════
# Run Single Algorithm
# ════════════════════════════════════════════════════════════════════════════

def run_single_algorithm_with_repeats(strategy_class, num_voters, num_rounds, num_projects, num_candidates, base_seed, repeats_per_dist):
    distributions = ['uniform', 'good_majority', 'bad_majority', 'shared_bias']
    results_by_distribution = {}
    raw_errors_by_distribution = {}
    for dist_idx, dist_name in enumerate(distributions):
        dist_errors = []
        for run in range(repeats_per_dist):
            strategy = strategy_class()
            seed = base_seed + dist_idx * 10000 + run * 1000
            system = VotingSystem(num_voters, strategy, num_projects, num_candidates, seed, error_distribution=dist_name)
            errors = system.run_multiple_rounds(num_rounds, verbose=False)
            dist_errors.append(errors)
        dist_errors = np.array(dist_errors)
        raw_errors_by_distribution[dist_name] = dist_errors
        avg_dist_errors = np.mean(dist_errors, axis=0)
        initial = avg_dist_errors[0]
        final = avg_dist_errors[-1]
        minimum = np.min(avg_dist_errors)
        improvement = initial - final
        improvement_percent = (improvement / initial * 100) if initial > 0 else 0
        results_by_distribution[dist_name] = {'initial': initial, 'final': final, 'min': minimum, 'improvement': improvement, 'improvement_percent': improvement_percent}
    avg_final = np.mean([results_by_distribution[dist]['final'] for dist in distributions])
    avg_initial = np.mean([results_by_distribution[dist]['initial'] for dist in distributions])
    avg_improvement_percent = np.mean([results_by_distribution[dist]['improvement_percent'] for dist in distributions])
    avg_min = np.mean([results_by_distribution[dist]['min'] for dist in distributions])
    return {'name': strategy_class().get_name(), 'final': avg_final, 'initial': avg_initial, 'min': avg_min, 'improvement': avg_initial - avg_final, 'improvement_percent': avg_improvement_percent, 'by_distribution': results_by_distribution,'raw_errors': raw_errors_by_distribution
}

# ════════════════════════════════════════════════════════════════════════════
# Plot Function - One plot per algorithm (averaged across all project counts)
# ════════════════════════════════════════════════════════════════════════════

def plot_error_over_rounds_aggregated(strategy_name, all_raw_errors_by_project):
    """
    all_raw_errors_by_project:
    dict -> num_projects : dict -> distribution name : np.array shape (repeats, num_rounds)
    """

    distributions = ['uniform', 'good_majority', 'bad_majority', 'shared_bias']
    num_rounds = all_raw_errors_by_project[PROJECT_COUNTS[0]][distributions[0]].shape[1]
    rounds = np.arange(num_rounds)

    plt.figure(figsize=(10, 6))

    # Collect all errors across all project counts for each distribution
    all_means = []
    for dist in distributions:
        # Average across all project counts for this distribution
        dist_means_across_projects = []
        for num_projects in PROJECT_COUNTS:
            mean_error = np.mean(all_raw_errors_by_project[num_projects][dist], axis=0)
            dist_means_across_projects.append(mean_error)

        # Average across all project counts
        avg_dist_mean = np.mean(np.array(dist_means_across_projects), axis=0)
        all_means.append(avg_dist_mean)
        plt.plot(rounds, avg_dist_mean, label=dist)

    # Global average across all distributions and project counts
    global_mean = np.mean(np.array(all_means), axis=0)
    plt.plot(rounds, global_mean, linewidth=3, linestyle='--', label='Global Average', color='black')

    plt.title(f'Error vs Voting Round – {strategy_name}\n(Averaged across {len(PROJECT_COUNTS)} project counts: {PROJECT_COUNTS})')
    plt.xlabel('Voting Round')
    plt.ylabel('Error')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

# ════════════════════════════════════════════════════════════════════════════
# Run Half Split Algorithms - MAIN EXECUTION
# ════════════════════════════════════════════════════════════════════════════

all_strategies = [
    HalfSplitMode1Strategy, HalfSplitMode2Strategy, HalfSplitMode3Strategy,
    HalfSplitMode4Strategy, HalfSplitMode5Strategy, HalfSplitMode6Strategy,
    HalfSplitMode7Strategy, HalfSplitMode8Strategy, HalfSplitMode9Strategy
]

total_algorithms = len(all_strategies)
total_runs = total_algorithms * len(PROJECT_COUNTS)
total_simulations = total_runs * REPEATS_PER_DISTRIBUTION * 4

print(f"\n{'='*130}")
print(f"🚀 HALF SPLIT ALGORITHMS")
print(f"🚀 Running {total_algorithms} algorithms × {len(PROJECT_COUNTS)} project counts × 4 distributions × {REPEATS_PER_DISTRIBUTION} repeats")
print(f"   📊 Total: {total_runs} main runs ({total_simulations} simulations!)")
print(f"   📊 Each simulation: {NUM_VOTERS} voters, {NUM_ROUNDS} rounds, {NUM_CANDIDATES} candidates")
print(f"   📈 Will create {total_algorithms} plots (one per algorithm, averaged across all project counts)")
print(f"{'='*130}\n")

start_total = time.time()
results_by_algorithm = {}
completed = 0

for algo_idx, strategy_class in enumerate(all_strategies, 1):
    strategy_name = strategy_class().get_name()
    results_by_algorithm[strategy_name] = {}

    # Store raw errors for all project counts
    all_raw_errors_by_project = {}

    print(f"\n┌{'─'*126}┐")
    print(f"│ [{algo_idx}/{total_algorithms}] 🔄 {strategy_name:<116} │")
    print(f"└{'─'*126}┘")

    for proj_idx, num_projects in enumerate(PROJECT_COUNTS, 1):
        start = time.time()
        print(f"   └─ [{proj_idx}/{len(PROJECT_COUNTS)}] {num_projects} Projects... ", end='', flush=True)

        results = run_single_algorithm_with_repeats(strategy_class, NUM_VOTERS, NUM_ROUNDS, num_projects, NUM_CANDIDATES, BASE_SEED, REPEATS_PER_DISTRIBUTION)
        results_by_algorithm[strategy_name][num_projects] = results

        # Store raw errors for later aggregation
        all_raw_errors_by_project[num_projects] = results['raw_errors']

        elapsed = time.time() - start
        completed += 1
        avg_time_per_run = (time.time() - start_total) / completed
        remaining_runs = total_runs - completed
        eta_seconds = avg_time_per_run * remaining_runs
        eta = str(timedelta(seconds=int(eta_seconds)))
        progress_pct = (completed / total_runs) * 100

        dist_results = [f"{dist[:3].upper()}:{results['by_distribution'][dist]['improvement_percent']:>5.1f}%" for dist in ['uniform', 'good_majority', 'bad_majority', 'shared_bias']]
        dist_str = " | ".join(dist_results)

        print(f"✅ {elapsed:>5.1f}s")
        print(f"      📊 Avg Error: {results['final']:.4f} | Avg Improvement: {results['improvement_percent']:>5.1f}%")
        print(f"      📈 By Distribution: {dist_str}")
        print(f"      ⏱️  Progress: {progress_pct:>5.1f}% | ETA: {eta}")

    # After all project counts are done, create ONE aggregated plot for this algorithm
    print(f"   📊 Creating aggregated plot for {strategy_name}...")
    plot_error_over_rounds_aggregated(strategy_name, all_raw_errors_by_project)


total_time = time.time() - start_total
print(f"\n{'='*130}")
print(f"✅ HALF SPLIT ALGORITHMS COMPLETED! Total Time: {str(timedelta(seconds=int(total_time)))}")
print(f"✅ Created {total_algorithms} plots (one per algorithm)")
print(f"{'='*130}\n")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from typing import List
from dataclasses import dataclass
from abc import ABC, abstractmethod
import time
from datetime import timedelta


plt.rcParams['font.family'] = 'DejaVu Sans'

# ╔══════════════════════════════════════════════════════════════════════════╗
# ║ 🎯 הגדרות - שנה רק כאן! 🎯 ║
# ╚══════════════════════════════════════════════════════════════════════════╝

NUM_VOTERS = 100
NUM_ROUNDS = 50
NUM_CANDIDATES = 300
PROJECT_COUNTS = [2, 3, 4, 5]
REPEATS_PER_DISTRIBUTION = 10
BASE_SEED = 42


@dataclass
class Voter:
    id: int
    error_rate: float
    weight: float = 1.0
    previous_vote: np.ndarray = None

class WeightUpdateStrategy(ABC):
    @abstractmethod
    def update_weights_batch(self, voters: List[Voter], votes: np.ndarray, current_result: np.ndarray, round_num: int) -> None:
        pass
    @abstractmethod
    def get_name(self) -> str:
        pass

# ════════════════════════════════════════════════════════════════════════════
# 9 Third Split Strategies
# ════════════════════════════════════════════════════════════════════════════

class ThirdSplitDynamicStrategy(WeightUpdateStrategy):
    def __init__(self, mode: int):
        self.mode = mode
        self.top_multiplier = float(mode + 1)
        self.mid_multiplier = 1.0
        self.bottom_multiplier = 1.0 / float(mode + 1)
    def update_weights_batch(self, voters, votes, current_result, round_num):
        distances = [(np.linalg.norm(votes[i] - current_result), i) for i in range(len(voters))]
        distances.sort(key=lambda x: x[0])
        n = len(voters)
        top_end = n // 3
        mid_end = (2 * n) // 3
        for rank, (_, idx) in enumerate(distances):
            voter = voters[idx]
            if rank < top_end:
                mult = self.top_multiplier
            elif rank < mid_end:
                mult = self.mid_multiplier
            else:
                mult = self.bottom_multiplier
            voter.weight *= mult
            voter.weight = float(np.clip(voter.weight, 0.01, 100.0))
    def get_name(self):
        return f"Third Split Mode {self.mode}"

class ThirdSplitMode1Strategy(ThirdSplitDynamicStrategy):
    def __init__(self): super().__init__(1)
class ThirdSplitMode2Strategy(ThirdSplitDynamicStrategy):
    def __init__(self): super().__init__(2)
class ThirdSplitMode3Strategy(ThirdSplitDynamicStrategy):
    def __init__(self): super().__init__(3)
class ThirdSplitMode4Strategy(ThirdSplitDynamicStrategy):
    def __init__(self): super().__init__(4)
class ThirdSplitMode5Strategy(ThirdSplitDynamicStrategy):
    def __init__(self): super().__init__(5)
class ThirdSplitMode6Strategy(ThirdSplitDynamicStrategy):
    def __init__(self): super().__init__(6)
class ThirdSplitMode7Strategy(ThirdSplitDynamicStrategy):
    def __init__(self): super().__init__(7)
class ThirdSplitMode8Strategy(ThirdSplitDynamicStrategy):
    def __init__(self): super().__init__(8)
class ThirdSplitMode9Strategy(ThirdSplitDynamicStrategy):
    def __init__(self): super().__init__(9)


# ════════════════════════════════════════════════════════════════════════════
# Voting System
# ════════════════════════════════════════════════════════════════════════════

class VotingSystem:
    def __init__(self, num_voters: int, weight_strategy: WeightUpdateStrategy, num_projects: int=3,
                 num_candidates: int=1000, random_seed: int=None, error_distribution: str='uniform'):
        if random_seed is not None:
            np.random.seed(random_seed)
        self.num_voters = num_voters
        self.num_projects = num_projects
        self.num_candidates = num_candidates
        self.true_allocation = self._generate_random_truth()
        self.weight_strategy = weight_strategy
        self.error_distribution = error_distribution
        self.shared_bias_direction = None
        if error_distribution == 'shared_bias':
            self.shared_bias_direction = self._generate_shared_bias()
        self.voters = self._initialize_voters()
        self.error_history = []

    def _generate_random_truth(self):
        vals = np.random.random(self.num_projects)
        return vals/np.sum(vals)

    def _generate_shared_bias(self):
        bias = np.random.normal(0, 1, self.num_projects)
        return bias / np.linalg.norm(bias)

    def _initialize_voters(self):
        voters = []
        if self.error_distribution == 'uniform':
            for i in range(self.num_voters):
                voters.append(Voter(i, np.random.uniform(0, 2)))
        elif self.error_distribution == 'good_majority':
            num_good = int(0.7 * self.num_voters)
            for i in range(num_good):
                voters.append(Voter(i, np.random.uniform(0, 0.5)))
            for i in range(num_good, self.num_voters):
                voters.append(Voter(i, np.random.uniform(0.5, 2.0)))
        elif self.error_distribution == 'bad_majority':
            num_bad = int(0.7 * self.num_voters)
            for i in range(num_bad):
                voters.append(Voter(i, np.random.uniform(1.5, 2.0)))
            for i in range(num_bad, self.num_voters):
                voters.append(Voter(i, np.random.uniform(0, 1.5)))
        elif self.error_distribution == 'shared_bias':
            for i in range(self.num_voters):
                voters.append(Voter(i, np.random.uniform(0, 2)))
        return voters

    def generate_vote_from_truth(self, voter: Voter, round_num: int):
        np.random.seed(voter.id * 12345 + round_num)
        if self.error_distribution == 'shared_bias':
            error_vector = self.shared_bias_direction * voter.error_rate
            vote = self.true_allocation + error_vector
            vote = np.abs(vote)
            return vote / np.sum(vote)
        if voter.previous_vote is None:
            rand = np.random.random(self.num_projects)
            rand = rand / np.sum(rand)
            mix_factor = voter.error_rate / 2.0
            vote = (1 - mix_factor) * self.true_allocation + mix_factor * rand
            vote = vote / np.sum(vote)
            voter.previous_vote = vote
            return vote
        candidates = []
        for i in range(self.num_candidates):
            rand = np.random.random(self.num_projects)
            rand = rand / np.sum(rand)
            mix_factor = voter.error_rate / 2.0
            vote_candidate = (1 - mix_factor) * self.true_allocation + mix_factor * rand
            candidates.append(vote_candidate / np.sum(vote_candidate))
        candidates = np.array(candidates)
        distances = np.array([np.sum(np.abs(candidate - voter.previous_vote)) for candidate in candidates])
        weights = 1.0 / (distances + 1e-10)
        probabilities = weights / np.sum(weights)
        idx = np.random.choice(len(candidates), p=probabilities)
        vote = candidates[idx].copy()
        voter.previous_vote = vote
        return vote

    def aggregate_votes(self, votes: np.ndarray, weights: np.ndarray):
        return np.sum(votes*weights[:,None], axis=0)/np.sum(weights)

    def run_voting_round(self, round_num: int):
        votes = np.array([self.generate_vote_from_truth(voter, round_num) for voter in self.voters])
        weights = np.array([voter.weight for voter in self.voters])
        result = self.aggregate_votes(votes, weights)
        error = np.sum(np.abs(result - self.true_allocation))
        self.weight_strategy.update_weights_batch(self.voters, votes, result, round_num)
        self.error_history.append(error)
        return error

    def run_multiple_rounds(self, num_rounds: int, verbose: bool = True):
        for round_num in range(num_rounds):
            self.run_voting_round(round_num)
        return self.error_history

# ════════════════════════════════════════════════════════════════════════════
# Run Single Algorithm
# ════════════════════════════════════════════════════════════════════════════

def run_single_algorithm_with_repeats(strategy_class, num_voters, num_rounds, num_projects, num_candidates, base_seed, repeats_per_dist):
    distributions = ['uniform', 'good_majority', 'bad_majority', 'shared_bias']
    results_by_distribution = {}
    raw_errors_by_distribution = {}
    for dist_idx, dist_name in enumerate(distributions):
        dist_errors = []
        for run in range(repeats_per_dist):
            strategy = strategy_class()
            seed = base_seed + dist_idx * 10000 + run * 1000
            system = VotingSystem(num_voters, strategy, num_projects, num_candidates, seed, error_distribution=dist_name)
            errors = system.run_multiple_rounds(num_rounds, verbose=False)
            dist_errors.append(errors)
        dist_errors = np.array(dist_errors)
        raw_errors_by_distribution[dist_name] = dist_errors
        avg_dist_errors = np.mean(dist_errors, axis=0)
        initial = avg_dist_errors[0]
        final = avg_dist_errors[-1]
        minimum = np.min(avg_dist_errors)
        improvement = initial - final
        improvement_percent = (improvement / initial * 100) if initial > 0 else 0
        results_by_distribution[dist_name] = {'initial': initial, 'final': final, 'min': minimum, 'improvement': improvement, 'improvement_percent': improvement_percent}
    avg_final = np.mean([results_by_distribution[dist]['final'] for dist in distributions])
    avg_initial = np.mean([results_by_distribution[dist]['initial'] for dist in distributions])
    avg_improvement_percent = np.mean([results_by_distribution[dist]['improvement_percent'] for dist in distributions])
    avg_min = np.mean([results_by_distribution[dist]['min'] for dist in distributions])
    return {'name': strategy_class().get_name(), 'final': avg_final, 'initial': avg_initial, 'min': avg_min, 'improvement': avg_initial - avg_final, 'improvement_percent': avg_improvement_percent, 'by_distribution': results_by_distribution,'raw_errors': raw_errors_by_distribution
}

# ════════════════════════════════════════════════════════════════════════════
# Plot Function - One plot per algorithm (averaged across all project counts)
# ════════════════════════════════════════════════════════════════════════════

def plot_error_over_rounds_aggregated(strategy_name, all_raw_errors_by_project):
    """
    all_raw_errors_by_project:
    dict -> num_projects : dict -> distribution name : np.array shape (repeats, num_rounds)
    """

    distributions = ['uniform', 'good_majority', 'bad_majority', 'shared_bias']
    num_rounds = all_raw_errors_by_project[PROJECT_COUNTS[0]][distributions[0]].shape[1]
    rounds = np.arange(num_rounds)

    plt.figure(figsize=(10, 6))

    # Collect all errors across all project counts for each distribution
    all_means = []
    for dist in distributions:
        # Average across all project counts for this distribution
        dist_means_across_projects = []
        for num_projects in PROJECT_COUNTS:
            mean_error = np.mean(all_raw_errors_by_project[num_projects][dist], axis=0)
            dist_means_across_projects.append(mean_error)

        # Average across all project counts
        avg_dist_mean = np.mean(np.array(dist_means_across_projects), axis=0)
        all_means.append(avg_dist_mean)
        plt.plot(rounds, avg_dist_mean, label=dist)

    # Global average across all distributions and project counts
    global_mean = np.mean(np.array(all_means), axis=0)
    plt.plot(rounds, global_mean, linewidth=3, linestyle='--', label='Global Average', color='black')

    plt.title(f'Error vs Voting Round – {strategy_name}\n(Averaged across {len(PROJECT_COUNTS)} project counts: {PROJECT_COUNTS})')
    plt.xlabel('Voting Round')
    plt.ylabel('Error')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

# ════════════════════════════════════════════════════════════════════════════
# Run Third Split Algorithms - MAIN EXECUTION
# ════════════════════════════════════════════════════════════════════════════

all_strategies = [
    ThirdSplitMode1Strategy, ThirdSplitMode2Strategy, ThirdSplitMode3Strategy,
    ThirdSplitMode4Strategy, ThirdSplitMode5Strategy, ThirdSplitMode6Strategy,
    ThirdSplitMode7Strategy, ThirdSplitMode8Strategy, ThirdSplitMode9Strategy
]

total_algorithms = len(all_strategies)
total_runs = total_algorithms * len(PROJECT_COUNTS)
total_simulations = total_runs * REPEATS_PER_DISTRIBUTION * 4

print(f"\n{'='*130}")
print(f"🚀 THIRD SPLIT ALGORITHMS")
print(f"🚀 Running {total_algorithms} algorithms × {len(PROJECT_COUNTS)} project counts × 4 distributions × {REPEATS_PER_DISTRIBUTION} repeats")
print(f"   📊 Total: {total_runs} main runs ({total_simulations} simulations!)")
print(f"   📊 Each simulation: {NUM_VOTERS} voters, {NUM_ROUNDS} rounds, {NUM_CANDIDATES} candidates")
print(f"   📈 Will create {total_algorithms} plots (one per algorithm, averaged across all project counts)")
print(f"{'='*130}\n")

start_total = time.time()
results_by_algorithm = {}
completed = 0

for algo_idx, strategy_class in enumerate(all_strategies, 1):
    strategy_name = strategy_class().get_name()
    results_by_algorithm[strategy_name] = {}

    # Store raw errors for all project counts
    all_raw_errors_by_project = {}

    print(f"\n┌{'─'*126}┐")
    print(f"│ [{algo_idx}/{total_algorithms}] 🔄 {strategy_name:<116} │")
    print(f"└{'─'*126}┘")

    for proj_idx, num_projects in enumerate(PROJECT_COUNTS, 1):
        start = time.time()
        print(f"   └─ [{proj_idx}/{len(PROJECT_COUNTS)}] {num_projects} Projects... ", end='', flush=True)

        results = run_single_algorithm_with_repeats(strategy_class, NUM_VOTERS, NUM_ROUNDS, num_projects, NUM_CANDIDATES, BASE_SEED, REPEATS_PER_DISTRIBUTION)
        results_by_algorithm[strategy_name][num_projects] = results

        # Store raw errors for later aggregation
        all_raw_errors_by_project[num_projects] = results['raw_errors']

        elapsed = time.time() - start
        completed += 1
        avg_time_per_run = (time.time() - start_total) / completed
        remaining_runs = total_runs - completed
        eta_seconds = avg_time_per_run * remaining_runs
        eta = str(timedelta(seconds=int(eta_seconds)))
        progress_pct = (completed / total_runs) * 100

        dist_results = [f"{dist[:3].upper()}:{results['by_distribution'][dist]['improvement_percent']:>5.1f}%" for dist in ['uniform', 'good_majority', 'bad_majority', 'shared_bias']]
        dist_str = " | ".join(dist_results)

        print(f"✅ {elapsed:>5.1f}s")
        print(f"      📊 Avg Error: {results['final']:.4f} | Avg Improvement: {results['improvement_percent']:>5.1f}%")
        print(f"      📈 By Distribution: {dist_str}")
        print(f"      ⏱️  Progress: {progress_pct:>5.1f}% | ETA: {eta}")

    # After all project counts are done, create ONE aggregated plot for this algorithm
    print(f"   📊 Creating aggregated plot for {strategy_name}...")
    plot_error_over_rounds_aggregated(strategy_name, all_raw_errors_by_project)


total_time = time.time() - start_total
print(f"\n{'='*130}")
print(f"✅ THIRD SPLIT ALGORITHMS COMPLETED! Total Time: {str(timedelta(seconds=int(total_time)))}")
print(f"✅ Created {total_algorithms} plots (one per algorithm)")
print(f"{'='*130}\n")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from typing import List
from dataclasses import dataclass
from abc import ABC, abstractmethod
import time
from datetime import timedelta


plt.rcParams['font.family'] = 'DejaVu Sans'

# ╔══════════════════════════════════════════════════════════════════════════╗
# ║ 🎯 הגדרות - שנה רק כאן! 🎯 ║
# ╚══════════════════════════════════════════════════════════════════════════╝

NUM_VOTERS = 100
NUM_ROUNDS = 50
NUM_CANDIDATES = 300
PROJECT_COUNTS = [2, 3, 4, 5]
REPEATS_PER_DISTRIBUTION = 10
BASE_SEED = 42


@dataclass
class Voter:
    id: int
    error_rate: float
    weight: float = 1.0
    previous_vote: np.ndarray = None

class WeightUpdateStrategy(ABC):
    @abstractmethod
    def update_weights_batch(self, voters: List[Voter], votes: np.ndarray, current_result: np.ndarray, round_num: int) -> None:
        pass
    @abstractmethod
    def get_name(self) -> str:
        pass

# ════════════════════════════════════════════════════════════════════════════
# 9 Fifth Split Strategies
# ════════════════════════════════════════════════════════════════════════════

class FifthSplitDynamicStrategy(WeightUpdateStrategy):
    def __init__(self, mode: int):
        self.mode = mode
        self.mult_1st = float(mode + 1)
        self.mult_2nd = (1.0 + float(mode + 1)) / 2.0
        self.mult_3rd = 1.0
        self.mult_4th = 1.0 / self.mult_2nd
        self.mult_5th = 1.0 / float(mode + 1)
    def update_weights_batch(self, voters, votes, current_result, round_num):
        distances = [(np.linalg.norm(votes[i] - current_result), i) for i in range(len(voters))]
        distances.sort(key=lambda x: x[0])
        n = len(voters)
        fifth_1 = n // 5
        fifth_2 = (2 * n) // 5
        fifth_3 = (3 * n) // 5
        fifth_4 = (4 * n) // 5
        for rank, (_, idx) in enumerate(distances):
            voter = voters[idx]
            if rank < fifth_1:
                mult = self.mult_1st
            elif rank < fifth_2:
                mult = self.mult_2nd
            elif rank < fifth_3:
                mult = self.mult_3rd
            elif rank < fifth_4:
                mult = self.mult_4th
            else:
                mult = self.mult_5th
            voter.weight = voter.weight * mult
            voter.weight = max(0.01, min(100.0, voter.weight))
    def get_name(self):
        return f"Fifth Split Mode {self.mode}"

class FifthSplitMode1Strategy(FifthSplitDynamicStrategy):
    def __init__(self): super().__init__(1)
class FifthSplitMode2Strategy(FifthSplitDynamicStrategy):
    def __init__(self): super().__init__(2)
class FifthSplitMode3Strategy(FifthSplitDynamicStrategy):
    def __init__(self): super().__init__(3)
class FifthSplitMode4Strategy(FifthSplitDynamicStrategy):
    def __init__(self): super().__init__(4)
class FifthSplitMode5Strategy(FifthSplitDynamicStrategy):
    def __init__(self): super().__init__(5)
class FifthSplitMode6Strategy(FifthSplitDynamicStrategy):
    def __init__(self): super().__init__(6)
class FifthSplitMode7Strategy(FifthSplitDynamicStrategy):
    def __init__(self): super().__init__(7)
class FifthSplitMode8Strategy(FifthSplitDynamicStrategy):
    def __init__(self): super().__init__(8)
class FifthSplitMode9Strategy(FifthSplitDynamicStrategy):
    def __init__(self): super().__init__(9)

# ════════════════════════════════════════════════════════════════════════════
# Voting System
# ════════════════════════════════════════════════════════════════════════════

class VotingSystem:
    def __init__(self, num_voters: int, weight_strategy: WeightUpdateStrategy, num_projects: int=3,
                 num_candidates: int=1000, random_seed: int=None, error_distribution: str='uniform'):
        if random_seed is not None:
            np.random.seed(random_seed)
        self.num_voters = num_voters
        self.num_projects = num_projects
        self.num_candidates = num_candidates
        self.true_allocation = self._generate_random_truth()
        self.weight_strategy = weight_strategy
        self.error_distribution = error_distribution
        self.shared_bias_direction = None
        if error_distribution == 'shared_bias':
            self.shared_bias_direction = self._generate_shared_bias()
        self.voters = self._initialize_voters()
        self.error_history = []

    def _generate_random_truth(self):
        vals = np.random.random(self.num_projects)
        return vals/np.sum(vals)

    def _generate_shared_bias(self):
        bias = np.random.normal(0, 1, self.num_projects)
        return bias / np.linalg.norm(bias)

    def _initialize_voters(self):
        voters = []
        if self.error_distribution == 'uniform':
            for i in range(self.num_voters):
                voters.append(Voter(i, np.random.uniform(0, 2)))
        elif self.error_distribution == 'good_majority':
            num_good = int(0.7 * self.num_voters)
            for i in range(num_good):
                voters.append(Voter(i, np.random.uniform(0, 0.5)))
            for i in range(num_good, self.num_voters):
                voters.append(Voter(i, np.random.uniform(0.5, 2.0)))
        elif self.error_distribution == 'bad_majority':
            num_bad = int(0.7 * self.num_voters)
            for i in range(num_bad):
                voters.append(Voter(i, np.random.uniform(1.5, 2.0)))
            for i in range(num_bad, self.num_voters):
                voters.append(Voter(i, np.random.uniform(0, 1.5)))
        elif self.error_distribution == 'shared_bias':
            for i in range(self.num_voters):
                voters.append(Voter(i, np.random.uniform(0, 2)))
        return voters

    def generate_vote_from_truth(self, voter: Voter, round_num: int):
        np.random.seed(voter.id * 12345 + round_num)
        if self.error_distribution == 'shared_bias':
            error_vector = self.shared_bias_direction * voter.error_rate
            vote = self.true_allocation + error_vector
            vote = np.abs(vote)
            return vote / np.sum(vote)
        if voter.previous_vote is None:
            rand = np.random.random(self.num_projects)
            rand = rand / np.sum(rand)
            mix_factor = voter.error_rate / 2.0
            vote = (1 - mix_factor) * self.true_allocation + mix_factor * rand
            vote = vote / np.sum(vote)
            voter.previous_vote = vote
            return vote
        candidates = []
        for i in range(self.num_candidates):
            rand = np.random.random(self.num_projects)
            rand = rand / np.sum(rand)
            mix_factor = voter.error_rate / 2.0
            vote_candidate = (1 - mix_factor) * self.true_allocation + mix_factor * rand
            candidates.append(vote_candidate / np.sum(vote_candidate))
        candidates = np.array(candidates)
        distances = np.array([np.sum(np.abs(candidate - voter.previous_vote)) for candidate in candidates])
        weights = 1.0 / (distances + 1e-10)
        probabilities = weights / np.sum(weights)
        idx = np.random.choice(len(candidates), p=probabilities)
        vote = candidates[idx].copy()
        voter.previous_vote = vote
        return vote

    def aggregate_votes(self, votes: np.ndarray, weights: np.ndarray):
        return np.sum(votes*weights[:,None], axis=0)/np.sum(weights)

    def run_voting_round(self, round_num: int):
        votes = np.array([self.generate_vote_from_truth(voter, round_num) for voter in self.voters])
        weights = np.array([voter.weight for voter in self.voters])
        result = self.aggregate_votes(votes, weights)
        error = np.sum(np.abs(result - self.true_allocation))
        self.weight_strategy.update_weights_batch(self.voters, votes, result, round_num)
        self.error_history.append(error)
        return error

    def run_multiple_rounds(self, num_rounds: int, verbose: bool = True):
        for round_num in range(num_rounds):
            self.run_voting_round(round_num)
        return self.error_history

# ════════════════════════════════════════════════════════════════════════════
# Run Single Algorithm
# ════════════════════════════════════════════════════════════════════════════

def run_single_algorithm_with_repeats(strategy_class, num_voters, num_rounds, num_projects, num_candidates, base_seed, repeats_per_dist):
    distributions = ['uniform', 'good_majority', 'bad_majority', 'shared_bias']
    results_by_distribution = {}
    raw_errors_by_distribution = {}
    for dist_idx, dist_name in enumerate(distributions):
        dist_errors = []
        for run in range(repeats_per_dist):
            strategy = strategy_class()
            seed = base_seed + dist_idx * 10000 + run * 1000
            system = VotingSystem(num_voters, strategy, num_projects, num_candidates, seed, error_distribution=dist_name)
            errors = system.run_multiple_rounds(num_rounds, verbose=False)
            dist_errors.append(errors)
        dist_errors = np.array(dist_errors)
        raw_errors_by_distribution[dist_name] = dist_errors
        avg_dist_errors = np.mean(dist_errors, axis=0)
        initial = avg_dist_errors[0]
        final = avg_dist_errors[-1]
        minimum = np.min(avg_dist_errors)
        improvement = initial - final
        improvement_percent = (improvement / initial * 100) if initial > 0 else 0
        results_by_distribution[dist_name] = {'initial': initial, 'final': final, 'min': minimum, 'improvement': improvement, 'improvement_percent': improvement_percent}
    avg_final = np.mean([results_by_distribution[dist]['final'] for dist in distributions])
    avg_initial = np.mean([results_by_distribution[dist]['initial'] for dist in distributions])
    avg_improvement_percent = np.mean([results_by_distribution[dist]['improvement_percent'] for dist in distributions])
    avg_min = np.mean([results_by_distribution[dist]['min'] for dist in distributions])
    return {'name': strategy_class().get_name(), 'final': avg_final, 'initial': avg_initial, 'min': avg_min, 'improvement': avg_initial - avg_final, 'improvement_percent': avg_improvement_percent, 'by_distribution': results_by_distribution,'raw_errors': raw_errors_by_distribution
}

# ════════════════════════════════════════════════════════════════════════════
# Plot Function - One plot per algorithm (averaged across all project counts)
# ════════════════════════════════════════════════════════════════════════════

def plot_error_over_rounds_aggregated(strategy_name, all_raw_errors_by_project):
    """
    all_raw_errors_by_project:
    dict -> num_projects : dict -> distribution name : np.array shape (repeats, num_rounds)
    """

    distributions = ['uniform', 'good_majority', 'bad_majority', 'shared_bias']
    num_rounds = all_raw_errors_by_project[PROJECT_COUNTS[0]][distributions[0]].shape[1]
    rounds = np.arange(num_rounds)

    plt.figure(figsize=(10, 6))

    # Collect all errors across all project counts for each distribution
    all_means = []
    for dist in distributions:
        # Average across all project counts for this distribution
        dist_means_across_projects = []
        for num_projects in PROJECT_COUNTS:
            mean_error = np.mean(all_raw_errors_by_project[num_projects][dist], axis=0)
            dist_means_across_projects.append(mean_error)

        # Average across all project counts
        avg_dist_mean = np.mean(np.array(dist_means_across_projects), axis=0)
        all_means.append(avg_dist_mean)
        plt.plot(rounds, avg_dist_mean, label=dist)

    # Global average across all distributions and project counts
    global_mean = np.mean(np.array(all_means), axis=0)
    plt.plot(rounds, global_mean, linewidth=3, linestyle='--', label='Global Average', color='black')

    plt.title(f'Error vs Voting Round – {strategy_name}\n(Averaged across {len(PROJECT_COUNTS)} project counts: {PROJECT_COUNTS})')
    plt.xlabel('Voting Round')
    plt.ylabel('Error')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

# ════════════════════════════════════════════════════════════════════════════
# Run Fifth Split Algorithms - MAIN EXECUTION
# ════════════════════════════════════════════════════════════════════════════

all_strategies = [
    FifthSplitMode1Strategy, FifthSplitMode2Strategy, FifthSplitMode3Strategy,
    FifthSplitMode4Strategy, FifthSplitMode5Strategy, FifthSplitMode6Strategy,
    FifthSplitMode7Strategy, FifthSplitMode8Strategy, FifthSplitMode9Strategy
]

total_algorithms = len(all_strategies)
total_runs = total_algorithms * len(PROJECT_COUNTS)
total_simulations = total_runs * REPEATS_PER_DISTRIBUTION * 4

print(f"\n{'='*130}")
print(f"🚀 FIFTH SPLIT ALGORITHMS")
print(f"🚀 Running {total_algorithms} algorithms × {len(PROJECT_COUNTS)} project counts × 4 distributions × {REPEATS_PER_DISTRIBUTION} repeats")
print(f"   📊 Total: {total_runs} main runs ({total_simulations} simulations!)")
print(f"   📊 Each simulation: {NUM_VOTERS} voters, {NUM_ROUNDS} rounds, {NUM_CANDIDATES} candidates")
print(f"   📈 Will create {total_algorithms} plots (one per algorithm, averaged across all project counts)")
print(f"{'='*130}\n")

start_total = time.time()
results_by_algorithm = {}
completed = 0

for algo_idx, strategy_class in enumerate(all_strategies, 1):
    strategy_name = strategy_class().get_name()
    results_by_algorithm[strategy_name] = {}

    # Store raw errors for all project counts
    all_raw_errors_by_project = {}

    print(f"\n┌{'─'*126}┐")
    print(f"│ [{algo_idx}/{total_algorithms}] 🔄 {strategy_name:<116} │")
    print(f"└{'─'*126}┘")

    for proj_idx, num_projects in enumerate(PROJECT_COUNTS, 1):
        start = time.time()
        print(f"   └─ [{proj_idx}/{len(PROJECT_COUNTS)}] {num_projects} Projects... ", end='', flush=True)

        results = run_single_algorithm_with_repeats(strategy_class, NUM_VOTERS, NUM_ROUNDS, num_projects, NUM_CANDIDATES, BASE_SEED, REPEATS_PER_DISTRIBUTION)
        results_by_algorithm[strategy_name][num_projects] = results

        # Store raw errors for later aggregation
        all_raw_errors_by_project[num_projects] = results['raw_errors']

        elapsed = time.time() - start
        completed += 1
        avg_time_per_run = (time.time() - start_total) / completed
        remaining_runs = total_runs - completed
        eta_seconds = avg_time_per_run * remaining_runs
        eta = str(timedelta(seconds=int(eta_seconds)))
        progress_pct = (completed / total_runs) * 100

        dist_results = [f"{dist[:3].upper()}:{results['by_distribution'][dist]['improvement_percent']:>5.1f}%" for dist in ['uniform', 'good_majority', 'bad_majority', 'shared_bias']]
        dist_str = " | ".join(dist_results)

        print(f"✅ {elapsed:>5.1f}s")
        print(f"      📊 Avg Error: {results['final']:.4f} | Avg Improvement: {results['improvement_percent']:>5.1f}%")
        print(f"      📈 By Distribution: {dist_str}")
        print(f"      ⏱️  Progress: {progress_pct:>5.1f}% | ETA: {eta}")

    # After all project counts are done, create ONE aggregated plot for this algorithm
    print(f"   📊 Creating aggregated plot for {strategy_name}...")
    plot_error_over_rounds_aggregated(strategy_name, all_raw_errors_by_project)


total_time = time.time() - start_total
print(f"\n{'='*130}")
print(f"✅ FIFTH SPLIT ALGORITHMS COMPLETED! Total Time: {str(timedelta(seconds=int(total_time)))}")
print(f"✅ Created {total_algorithms} plots (one per algorithm)")
print(f"{'='*130}\n")
```

```python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
from typing import List
from dataclasses import dataclass
from abc import ABC, abstractmethod
import time
from datetime import timedelta

# הדבק כאן את כל ה-classes המקוריים
# Voter, WeightUpdateStrategy, VotingSystem

NUM_VOTERS = 100
NUM_ROUNDS = 50
NUM_CANDIDATES = 300
PROJECT_COUNTS = [2, 3, 4, 5]
REPEATS_PER_DISTRIBUTION = 10
BASE_SEED = 42


@dataclass
class Voter:
    id: int
    error_rate: float
    weight: float = 1.0
    previous_vote: np.ndarray = None

class WeightUpdateStrategy(ABC):
    @abstractmethod
    def update_weights_batch(self, voters: List[Voter], votes: np.ndarray, current_result: np.ndarray, round_num: int) -> None:
        pass
    @abstractmethod
    def get_name(self) -> str:
        pass

# ════════════════════════════════════════════════════════════════════════════
# 10 Linear Ranking Strategies
# ════════════════════════════════════════════════════════════════════════════

class LinearRankingRewardStrategy(WeightUpdateStrategy):
    def __init__(self, max_bonus_percent: float = 10.0):
        self.max_bonus_percent = max_bonus_percent
    def update_weights_batch(self, voters, votes, current_result, round_num):
        distances = [(np.linalg.norm(votes[i] - current_result), i, voters[i]) for i in range(len(voters))]
        distances.sort(key=lambda x: x[0])
        num_voters = len(voters)
        for rank, (dist, idx, voter) in enumerate(distances):
            if num_voters > 1:
                bonus_percent = self.max_bonus_percent * (num_voters - rank - 1) / (num_voters - 1)
            else:
                bonus_percent = self.max_bonus_percent
            multiplier = 1.0 + (bonus_percent / 100.0)
            voter.weight = voter.weight * multiplier
            voter.weight = max(0.01, min(100.0, voter.weight))
    def get_name(self):
        return f"Linear Ranking {self.max_bonus_percent:.0f}%"

class LinearRanking10Strategy(LinearRankingRewardStrategy):
    def __init__(self): super().__init__(10.0)
class LinearRanking20Strategy(LinearRankingRewardStrategy):
    def __init__(self): super().__init__(20.0)
class LinearRanking30Strategy(LinearRankingRewardStrategy):
    def __init__(self): super().__init__(30.0)
class LinearRanking40Strategy(LinearRankingRewardStrategy):
    def __init__(self): super().__init__(40.0)
class LinearRanking50Strategy(LinearRankingRewardStrategy):
    def __init__(self): super().__init__(50.0)
class LinearRanking60Strategy(LinearRankingRewardStrategy):
    def __init__(self): super().__init__(60.0)
class LinearRanking70Strategy(LinearRankingRewardStrategy):
    def __init__(self): super().__init__(70.0)
class LinearRanking80Strategy(LinearRankingRewardStrategy):
    def __init__(self): super().__init__(80.0)
class LinearRanking90Strategy(LinearRankingRewardStrategy):
    def __init__(self): super().__init__(90.0)
class LinearRanking100Strategy(LinearRankingRewardStrategy):
    def __init__(self): super().__init__(100.0)

# ════════════════════════════════════════════════════════════════════════════
# Voting System
# ════════════════════════════════════════════════════════════════════════════

class VotingSystem:
    def __init__(self, num_voters: int, weight_strategy: WeightUpdateStrategy, num_projects: int=3,
                 num_candidates: int=1000, random_seed: int=None, error_distribution: str='uniform'):
        if random_seed is not None:
            np.random.seed(random_seed)
        self.num_voters = num_voters
        self.num_projects = num_projects
        self.num_candidates = num_candidates
        self.true_allocation = self._generate_random_truth()
        self.weight_strategy = weight_strategy
        self.error_distribution = error_distribution
        self.shared_bias_direction = None
        if error_distribution == 'shared_bias':
            self.shared_bias_direction = self._generate_shared_bias()
        self.voters = self._initialize_voters()
        self.error_history = []

    def _generate_random_truth(self):
        vals = np.random.random(self.num_projects)
        return vals/np.sum(vals)

    def _generate_shared_bias(self):
        bias = np.random.normal(0, 1, self.num_projects)
        return bias / np.linalg.norm(bias)

    def _initialize_voters(self):
        voters = []
        if self.error_distribution == 'uniform':
            for i in range(self.num_voters):
                voters.append(Voter(i, np.random.uniform(0, 2)))
        elif self.error_distribution == 'good_majority':
            num_good = int(0.7 * self.num_voters)
            for i in range(num_good):
                voters.append(Voter(i, np.random.uniform(0, 0.5)))
            for i in range(num_good, self.num_voters):
                voters.append(Voter(i, np.random.uniform(0.5, 2.0)))
        elif self.error_distribution == 'bad_majority':
            num_bad = int(0.7 * self.num_voters)
            for i in range(num_bad):
                voters.append(Voter(i, np.random.uniform(1.5, 2.0)))
            for i in range(num_bad, self.num_voters):
                voters.append(Voter(i, np.random.uniform(0, 1.5)))
        elif self.error_distribution == 'shared_bias':
            for i in range(self.num_voters):
                voters.append(Voter(i, np.random.uniform(0, 2)))
        return voters

    def generate_vote_from_truth(self, voter: Voter, round_num: int):
        np.random.seed(voter.id * 12345 + round_num)
        if self.error_distribution == 'shared_bias':
            error_vector = self.shared_bias_direction * voter.error_rate
            vote = self.true_allocation + error_vector
            vote = np.abs(vote)
            return vote / np.sum(vote)
        if voter.previous_vote is None:
            rand = np.random.random(self.num_projects)
            rand = rand / np.sum(rand)
            mix_factor = voter.error_rate / 2.0
            vote = (1 - mix_factor) * self.true_allocation + mix_factor * rand
            vote = vote / np.sum(vote)
            voter.previous_vote = vote
            return vote
        candidates = []
        for i in range(self.num_candidates):
            rand = np.random.random(self.num_projects)
            rand = rand / np.sum(rand)
            mix_factor = voter.error_rate / 2.0
            vote_candidate = (1 - mix_factor) * self.true_allocation + mix_factor * rand
            candidates.append(vote_candidate / np.sum(vote_candidate))
        candidates = np.array(candidates)
        distances = np.array([np.sum(np.abs(candidate - voter.previous_vote)) for candidate in candidates])
        weights = 1.0 / (distances + 1e-10)
        probabilities = weights / np.sum(weights)
        idx = np.random.choice(len(candidates), p=probabilities)
        vote = candidates[idx].copy()
        voter.previous_vote = vote
        return vote

    def aggregate_votes(self, votes: np.ndarray, weights: np.ndarray):
        return np.sum(votes*weights[:,None], axis=0)/np.sum(weights)

    def run_voting_round(self, round_num: int):
        votes = np.array([self.generate_vote_from_truth(voter, round_num) for voter in self.voters])
        weights = np.array([voter.weight for voter in self.voters])
        result = self.aggregate_votes(votes, weights)
        error = np.sum(np.abs(result - self.true_allocation))
        self.weight_strategy.update_weights_batch(self.voters, votes, result, round_num)
        self.error_history.append(error)
        return error

    def run_multiple_rounds(self, num_rounds: int, verbose: bool = True):
        for round_num in range(num_rounds):
            self.run_voting_round(round_num)
        return self.error_history

class NoUpdateStrategy(WeightUpdateStrategy):
    def update_weights_batch(self, voters, votes, current_result, round_num): pass
    def get_name(self): return "No Update"

distributions = ['uniform', 'good_majority', 'bad_majority', 'shared_bias']

dist_means = []
for dist_idx, dist_name in enumerate(distributions):
    project_means = []
    for num_projects in PROJECT_COUNTS:
        run_errors = []
        for run in range(REPEATS_PER_DISTRIBUTION):
            seed = BASE_SEED + dist_idx * 10000 + run * 1000
            system = VotingSystem(NUM_VOTERS, NoUpdateStrategy(), num_projects, NUM_CANDIDATES, seed, dist_name)
            error = system.run_voting_round(0)
            run_errors.append(error)
        project_means.append(np.mean(run_errors))
    dist_means.append(np.mean(project_means))
    print(f"{dist_name}: {np.mean(project_means):.4f}")

print(f"\n➡️  Global Average בסיבוב 0: {np.mean(dist_means):.4f}")
```
